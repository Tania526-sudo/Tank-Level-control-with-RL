## 1. Load Data and Q-table

```python
import numpy as np
import matplotlib.pyplot as plt

# Load Q-table and set parameters
Q = np.loadtxt('Q_Mattrix.txt')
print('Q-table shape:', Q.shape)
```

## 2. Q-table Heatmap

```python
import seaborn as sns
plt.figure(figsize=(12, 6))
sns.heatmap(Q, cmap='viridis')
plt.title('Q-table Heatmap (States vs Actions)')
plt.xlabel('Actions')
plt.ylabel('States')
plt.show()
```

## 3. Visualizing Training Rewards

```python
# Load training reward log (optional: you could also log it to file during training)
# Here we simulate it for demonstration
episode_rewards = np.loadtxt('Q_Mattrix.txt').sum(axis=1)  # not real rewards; simulated

plt.figure(figsize=(10, 4))
plt.plot(episode_rewards)
plt.title('Simulated Training Reward per State')
plt.xlabel('State')
plt.ylabel('Estimated Cumulative Q-value')
plt.grid(True)
plt.show()
```

## 4. Most Optimal Actions per State

```python
# Determine optimal action per state
optimal_actions = np.argmax(Q, axis=1)

plt.figure(figsize=(10, 4))
plt.plot(optimal_actions, marker='o')
plt.title('Optimal Action per State')
plt.xlabel('State Index')
plt.ylabel('Best Action Index')
plt.grid(True)
plt.show()
```
