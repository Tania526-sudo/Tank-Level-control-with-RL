```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import collections
```

```python
import numpy as np


tank_levels = np.random.normal(loc=5.0, scale=0.3, size=100)
rewards = np.cumsum(np.random.randn(50))
actions = np.random.choice([0, 1], size=100)


q_table = np.random.rand(5, 2)
```

```python
def plot_tank_levels(tank_levels, target_level):
    plt.figure(figsize=(10, 4))
    plt.plot(range(len(tank_levels)), tank_levels, label='Tank Level', color='blue')
    plt.axhline(target_level, color='green', linestyle='--', label='Target Level')
    plt.xlabel('Timestep')
    plt.ylabel('Water Level')
    plt.title('Tank Level Over Time')
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```

```python
def plot_episode_rewards(rewards):
    plt.figure(figsize=(8, 4))
    plt.plot(range(len(rewards)), rewards, label='Total Reward per Episode', color='orange')
    plt.xlabel('Episode')
    plt.ylabel('Cumulative Reward')
    plt.title('Reward Evolution')
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```

```python
def plot_q_table(q_table):
    if isinstance(q_table, dict):
        max_state = max(s for s, _ in q_table.keys())
        max_action = max(a for _, a in q_table.keys())
        q_matrix = np.zeros((max_state + 1, max_action + 1))
        for (s, a), value in q_table.items():
            q_matrix[s, a] = value
    else:
        q_matrix = q_table  # already numpy array

    plt.figure(figsize=(10, 6))
    sns.heatmap(q_matrix, annot=True, fmt=".2f", cmap="viridis", cbar_kws={'label': 'Q-value'})
    plt.xlabel('Action')
    plt.ylabel('State')
    plt.title('Q-table Heatmap')
    plt.tight_layout()
    plt.show()
```

```python
def plot_action_distribution(actions):
    action_counts = collections.Counter(actions)
    plt.bar(action_counts.keys(), action_counts.values(), color='teal')
    plt.xlabel('Action')
    plt.ylabel('Frequency')
    plt.title('Action Distribution Over Episodes')
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```

```python
from IPython.display import clear_output
import time

def animate_tank_level(tank_levels):
    for t in range(len(tank_levels)):
        clear_output(wait=True)
        print(f"Step {t}: Tank Level = {tank_levels[t]:.2f}")
        time.sleep(0.2)
```

```python
plot_tank_levels(tank_levels, target_level=5)
```

```python
plot_episode_rewards(rewards)
```

```python
plot_q_table(q_table)
```

```python
plot_action_distribution(actions)
```
